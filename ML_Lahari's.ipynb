```python
#Install Required Libraries
!pip install nltk
!pip install scikit-learn

#Import Libraries
import nltk
import random
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')

#Creating Own Dataset

questions = [
    "hello",
    "hi",
    "my name",
    "when are exams",
    "what is machine learning",
    "who is hod",
    "what subjects are in ml",
    "bye"
]

answers = [
    "Hello! How can I help you?",
    "Hi there!",
    "Lahari...",
    "Exams start from May 10.",
    "Machine Learning is a branch of AI that allows systems to learn from data.",
    "The HOD of CSE department is Dr. Ramesh.",
    "ML subjects include regression, classification, clustering and neural networks.",
    "Goodbye!"
]

#Training NLP Model

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(questions)

#Chatbot Function
def chatbot(user_input):

    user_vec = vectorizer.transform([user_input])

    similarity = cosine_similarity(user_vec, X)

    index = np.argmax(similarity)

    return answers[index]

#Running the Chatbot

while True:

    user = input("You: ")

    if user.lower() == "exit":
        print("Bot: Goodbye!")
        break

    response = chatbot(user)

    print("Bot:", response)
```
